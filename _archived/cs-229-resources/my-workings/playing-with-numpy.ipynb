### Numpy Arrays

```python
first_array = np.array([1,2,3,4,5,6])
dataType = first_array.dtype
print(f"Data type of the array is `{dataType}`")

numberOfDimensions = first_array.ndim
print(f"Number of dimensions of the array is `{numberOfDimensions}`")

evens = [i for i in range(2, 12, 2)]
odds = [j for j in range(1, 11, 2)]

second_array = np.array([evens, odds])
print(second_array.ndim)
```

```python
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(0)

# Generate random data for linear regression
num_samples = 250
X = 3 * np.random.rand(num_samples, 1)  # Random feature (1D in this case)

# Introduce randomness to the slope and intercept
true_slope = 4 + np.random.randn()  # Random slope
true_intercept = 5 + np.random.randn()  # Random intercept

y = true_intercept + true_slope * X + 2 * np.random.randn(num_samples, 1)  # Linear relationship with added noise

# Plot the generated dataset
plt.figure(figsize=(8, 8))
plt.scatter(X, y, color='blue', label='Generated Data')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Random Dataset for Linear Regression')
plt.legend()
plt.grid(True)

plt.ylim(0, None)
plt.show()

data = np.hstack((X, y))
np.savetxt('dummy_linreg.csv', data, delimiter=',', header='Dependent Variable,Independent Variable')
```
