```python
import numpy as np
import matplotlib.pyplot as plt

# Define a grid in 3D space
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)


# Define a 3D parabolic function
z = 2 * x**2 + 3 * y**2

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D surface
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the title of the plot
plt.title('3D Parabolic Function')

# Show the plot
plt.show()
```
