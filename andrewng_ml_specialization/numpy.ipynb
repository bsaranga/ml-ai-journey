```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
x = np.linspace(0, 10, n+1)
y = np.e**(-x/10)*np.sin(x)

plt.plot(x, y)
```

```python
filter = np.vectorize(lambda x : x >=4 and x <= 7)
mean = np.mean(y[filter(x)])
std = np.std(y[filter(x)])
print(f"Mean: {mean}, Std: {std}")
```

```python
percentile = np.percentile(y[filter(x)], 80)
print(f"80th Percentile: {percentile}")
```

```python
dydx = np.gradient(y, x)
plt.plot(x, dydx, label='Derivative')
zeroes = x[1:][dydx[1:] * dydx[:-1] < 0]
plt.scatter(zeroes, np.zeros_like(zeroes), color='red', label='Zeroes')
```

```python
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)

xv, yv = np.meshgrid(x, y)
z = xv**2 + yv**2

plt.contourf(xv, yv, z, levels=30)
plt.colorbar(label='Function Value')
```
