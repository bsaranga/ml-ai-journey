```python
import numpy as np
import matplotlib.pyplot as plt

w = 1
b = 0

fix, ax = plt.subplots(figsize=(8, 6))
x = np.linspace(-10, 10, 100)
y = 1/ (1 + np.exp(-(x*w + b)))
y_hat = w*x + b
ax.plot(x, y, label='Sigmoid Function', color='blue')
ax.plot(x, y_hat, label='Linear Function', color='green', linestyle='--')
ax.set_title('Sigmoid Function')
ax.axhline(0, color='black', lw=0.5, ls='-')
ax.axvline(0, color='black', lw=0.5, ls='-')
ax.axhline(0.5, color='red', lw=0.5, ls='--', label='y=0.5')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

ax.set_xlim(-10, 10)
ax.set_ylim(-0.1, 1.1)

ax.legend()
```
